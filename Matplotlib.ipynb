# Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 1000)
y = np.sin(x)
plt.plot(x, y);
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x[:50], y[:50], color='green')
```

```python
plt.plot(x,y,color='b')
plt.plot(x, np.cos(x), color='g')
```
